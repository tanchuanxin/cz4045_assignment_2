```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Tied Weights vs No Tied Weights

```python
fnn_nt_df = pd.read_csv('../logs/model-FNN-not_tied.csv')
fnn_nt_df['perpexity'].plot()

fnn_t_df = pd.read_csv('../logs/model-FNN-tied.csv')
fnn_t_df['perpexity'].plot()

plt.title('FNN Comparison (Tied Weights vs No Tied Weights)')
plt.xlabel('Epochs')
plt.ylabel('Perplexity')
plt.legend(['No Tied Weights', 'Tied Weights'])
```

# Adam vs RMSprop

```python
fnn_nt_df = pd.read_csv('../logs/model-FNN-not_tied.csv')
fnn_nt_df['perpexity'].plot()

rms_fnn_nt_df = pd.read_csv('../logs/model-rms-FNN-not_tied.csv')
rms_fnn_nt_df['perpexity'].plot()

plt.title('FNN Comparison with no tied weights (Adam vs RMSprop)')
plt.xlabel('Epochs')
plt.ylabel('Perplexity')
plt.legend(['Adam', 'RMS'])
```

```python
fnn_nt_df = pd.read_csv('../logs/model-FNN-tied.csv')
fnn_nt_df['perpexity'].plot()

rms_fnn_nt_df = pd.read_csv('../logs/model-rms-FNN-tied.csv')
rms_fnn_nt_df['perpexity'].plot()

plt.title('FNN Comparison with tied weights (Adam vs RMSprop)')
plt.xlabel('Epochs')
plt.ylabel('Perplexity')
plt.legend(['Adam', 'RMS'])
```

# Model Comparison (FNN, LSM, Transformer) with Adam Optimizer

```python
fnn_t_df = pd.read_csv('../logs/model-FNN-tied.csv')
fnn_t_df['perpexity'].plot()

lstm_t_df = pd.read_csv('../logs/model-LSTM-tied.csv')
lstm_t_df['perpexity'].plot()

tran_t_df = pd.read_csv('../logs/model-Transformer-tied.csv')
tran_t_df['perpexity'].plot()

plt.title('Model Comparison (Tied Weights)')
plt.xlabel('Epochs')
plt.ylabel('Perplexity')
plt.legend(['FNN', 'LSTM', 'Transformers'])
```
